## Mi Estrategia para el Blackjack 

En mi versión de Blackjack, quiero simular un juego simple pero realista de 1 jugador contra el dealer en la consola. Primero construyo un mazo estándar de 52 cartas usando Python. Los rangos de las cartas y sus valores fijos (2–10, J=10, Q=10, K=10, A=11) los guardo en una tupla, porque estos valores nunca cambian durante el juego. Luego creo el mazo completo como una lista, repitiendo los rangos para los cuatro palos, de modo que el mazo pueda barajarse y modificarse mientras se juega.

Al inicio de cada ronda, barajo el mazo y reparto dos cartas para mí y dos para el dealer, eliminando las cartas directamente de la lista. Tanto mi mano como la del dealer las guardo en listas, porque necesito seguir agregando cartas con .append() cada vez que alguien pide una carta (“hit”). Para calcular el valor total de una mano, recorro la lista de cartas y sumo sus valores usando la tupla donde guardé los valores fijos.

Durante mi turno, el programa me muestra mi mano y mi total, y luego me pregunta en la consola si quiero "hit" (pedir carta) o "stand" (quedarme) usando input(). Uso un ciclo while que continúa mientras yo elija pedir carta y no me haya pasado de 21. Para manejar el problema del As (que puede valer 11 o 1), primero cuento todos los Ases como 11 y luego uso otro ciclo while que reduce algunos Ases de 11 a 1 si mi total supera 21.

En el turno del dealer, su lógica es automática: el dealer debe seguir pidiendo cartas hasta que su total sea al menos 17. Para eso uso otro ciclo while que continúa repartiendo cartas mientras el total del dealer sea menor a 17, ajustando los Ases si hace falta.
Al final de la ronda, comparo los totales con condiciones if/elif para decidir si yo gano, si gana el dealer, si alguien se pasó (bust) o si hay un empate (push). Finalmente, el programa me pregunta si quiero jugar otra ronda, y utilizo un bucle para repetir todo el proceso hasta que yo decida detenerme.
